## Udemy
##### General Info on Udemy:
 Udemy, Inc. is an American massive open online course provider aimed at professional adults and students.           

 It was founded in May 2010 by Eren Bali, Gagan Biyani, and Oktay Caglar.

 As of Jan 2020, the platform has more than 35 million students and 57,000 instructors teaching courses in over 65 languages.

The project contains dataset with entire details of various courses from sizeable subjects from the year 2011 to 2017 availalable on Udemy platform. this dataset contains 12 columns and 3.682 records of courses from 4 subjects (Business Finance, Graphic Design, Musical Instruments and Web Design) taken from Udemy.


##### Features:
•	course_id - id field for courses

•	course_title - title field for courses

•	url - url field to course page

•	is_paid - boolean field - True for Paid / False for Free

•	price - price field for course fee

•	num_subscribers - demand field for each courses

•	num_reviews - review number for each courses

•	num_lectures - lecture per course

•	level - course level by trainee experience

•	content_duration - course duration in hours

•	published_timestamp - timestamp field for publication

•	subject - course type field

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('udemy_courses.csv')
```

```python
df
```

### Preparing the Data

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

> The data set has no null values. but it includes a duplicated rows so we will drop them.

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
for x in df.index:
    if (df.loc[x, 'num_lectures'] == 0):
        df.drop(x, inplace = True)
```

> drop the courses that has not lectures as it not make sense for the courses to not contain any lecture.

```python
df['course_id'].nunique()
```

> we have two unique and unneeded columns( 'course_id' and 'url') so we will drop them.

```python
df.drop(columns=['course_id', 'url'], inplace=True)
```

```python
df.shape
```

```python
#changing the 'published_timestamp' to correct Datatye
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df['year'] = df['published_timestamp'].dt.year
```

```python
df
```

## EDA and Visualization

### 1- HeatMap: 
**To illustrate the colleration between the features.**

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

> We can conclude from the HeatMap that 'num_subscribers' column has a negative colleration with 'is_paid' which mean that free courses have more subscribers than paid ones. we can also see that 'price' has a positive colleration with 'num_lectures'. 

## Number of Paid and Free courses

```python
sns.countplot(x='is_paid',data=df)
plt.title('Paid and Free courses')
plt.show()
```

```python
plt.pie(df['is_paid'].value_counts(),labels=('Paid','Free'),autopct ='%1.2f%%',shadow = True, explode = [0,0.15])
plt.title('Paid and Free courses')
plt.axis('equal')
plt.show()
```

> We can see that most of the courses are paid. But which one are on demand?

### Relation between paid courses and number of subscribers

```python
plt.figure(figsize=(8,5))
plt.title('Average Demand (Free/Paid)')
sns.barplot(x='is_paid', y='num_subscribers', data=df)
```

> Free courses are in demand more than paid courses.

### Number of courses in each subject

```python
plt.figure(figsize=(8,5))
sns.countplot(data = df, x = 'subject', order = df.subject.value_counts().index)
```

```python
plt.pie(df['subject'].value_counts(),labels=df['subject'].value_counts().index,autopct ='%1.2f%%',shadow = True, explode = [0.1,0,0,0])
plt.title('Number of courses in each subject')
plt.axis('equal')
plt.show()
```

> 'Web Development' and 'Business Finance' have the bigger number of courses and they are almost equal. But which subject is most popular to the subscribers? 

### Which subject has the max number of Subscribers?

```python
plt.figure(figsize=(8,5))
plt.title('Average Demand by Course Subjects')
sns.barplot(x='subject', y='num_subscribers', data=df)
```

```python
plt.figure(figsize=(8,5))
plt.title('Average Demand by Course Subjects')
sns.barplot(x='subject', y='num_subscribers',hue='is_paid', data=df)
```

> it is obvious that web development and Free courses are the most popular to the subscribers.

## Number of courses in each level

```python
plt.figure(figsize=(8,5))
sns.countplot(data = df, x = 'level', order = df.level.value_counts().index)
```

```python
plt.figure(figsize=(7,7))
plt.pie(df.level.value_counts(), labels = ['All Levels', 'Beginner', 'Intermediate', 'Expert'], autopct ='%1.f%%', explode = [0, 0, 0, 0.2])
plt.title("The percentage of each level")
plt.axis('equal')
plt.show()
```

> We can conclude that the bigger number of courses are in all levels and only few courses are in expert level.

## Relation between price and number of subscribers

```python
sns.scatterplot(x='price',y='num_subscribers',data=df)
plt.show()
```

> The most popular courses tend to be either free or very expensive.

## Which course duration and number of lectures does subscribers prefer?

```python
sns.scatterplot(x='num_lectures',y='num_subscribers',data=df)
plt.show()
```

```python
sns.scatterplot(x='content_duration',y='num_subscribers',data=df)
plt.show()
```

> Plots with the number of lectures and content duration show that most popular courses are shorter, and have fewer number of content hours.

### Which level of subjects are among highly priced than others?

```python
plt.figure(figsize=(12,6))
sns.barplot(x='subject',y='price',hue='level',data=df)
plt.title('Price comparison based on the subject and level of difficulty')
plt.show()
```

> it can be clearly seen that on most of the subjects Expert level of courses are seemed to be highly priced compared to other level (other than for Musical Instruments).The most highly priced course is seemed to be under Web Development subject.

## Total price of courses over the years

```python
plt.plot(df.groupby('year').sum()['price'],ls = '-',lw = 3, alpha = 0.85)  
plt.title(" Total price of courses over the years ", size=15 )
plt.xlabel('Years',color = 'k',size=12)
plt.ylabel('price',color = 'k',size=12)
plt.show()
```
